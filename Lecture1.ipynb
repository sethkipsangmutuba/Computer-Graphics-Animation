```python
# =========================================================
# 1. SETUP
# =========================================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D

plt.style.use('seaborn-v0_8')


# =========================================================
# 2. GENERATE ALL DATA FIRST
# =========================================================

# Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Simulation data
t = np.linspace(0, 10, 200)
sim_x = np.sin(t)
sim_y = np.cos(t)

# Spatial data (Kenya-like bounds)
lon = np.random.uniform(34, 42, 200)
lat = np.random.uniform(-5, 5, 200)
values = np.random.rand(200)

# Parametric CAD data
theta = np.linspace(0, 2*np.pi, 200)
r = 5
cad_x = r * np.cos(theta)
cad_y = r * np.sin(theta)

# Image data
from sklearn.datasets import load_sample_image
image = load_sample_image("china.jpg")


# =========================================================
# 3. CREATE SUBPLOT GRID (3 PER ROW)
# =========================================================

fig = plt.figure(figsize=(15, 10))


# ---------------------------------------------------------
# (1) 2D Data Mapping
# ---------------------------------------------------------
ax1 = fig.add_subplot(2, 3, 1)
ax1.scatter(X[:, 0], X[:, 1], c=y)
ax1.set_title("2D Mapping")
ax1.set_xlabel("Sepal Length")
ax1.set_ylabel("Sepal Width")


# ---------------------------------------------------------
# (2) 3D Scientific Visualization
# ---------------------------------------------------------
ax2 = fig.add_subplot(2, 3, 2, projection='3d')
ax2.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)
ax2.set_title("3D Visualization")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_zlabel("Z")


# ---------------------------------------------------------
# (3) Simulation
# ---------------------------------------------------------
ax3 = fig.add_subplot(2, 3, 3)
ax3.plot(sim_x, sim_y)
ax3.set_title("Simulation (Motion)")
ax3.set_xlabel("x")
ax3.set_ylabel("y")
ax3.axis('equal')


# ---------------------------------------------------------
# (4) Spatial / Cartography
# ---------------------------------------------------------
ax4 = fig.add_subplot(2, 3, 4)
ax4.scatter(lon, lat, c=values)
ax4.set_title("Spatial Mapping")
ax4.set_xlabel("Longitude")
ax4.set_ylabel("Latitude")


# ---------------------------------------------------------
# (5) CAD (Parametric Geometry)
# ---------------------------------------------------------
ax5 = fig.add_subplot(2, 3, 5)
ax5.plot(cad_x, cad_y)
ax5.set_title("Parametric Design")
ax5.set_aspect('equal')


# ---------------------------------------------------------
# (6) Multimedia (Image)
# ---------------------------------------------------------
ax6 = fig.add_subplot(2, 3, 6)
ax6.imshow(image)
ax6.set_title("Image Display")
ax6.axis('off')


# =========================================================
# 4. FINAL LAYOUT
# =========================================================

plt.tight_layout()
plt.show()
```

# 1.2 A Brief History of Computer Graphics

This section focuses on the historical evolution of computer graphics, with emphasis on foundational developments that continue to influence modern systems.

---

## Key Orientation

- Emphasis on **fundamental principles and techniques** that remain relevant  
- **Hardware evolution** is prioritized because it has historically driven software development and capabilities  

---

## Early Developments

### Hardcopy Graphics

Initial graphics output was limited to:

- Teletypes  
- Line printers  

**Characteristics:**

- Non-interactive  
- Low fidelity  

---

### CRT-Based Systems (1950s)

Early breakthrough: **Cathode Ray Tube (CRT) displays**

**Example:**
- Whirlwind Computer (MIT, ~1950)

**Enabled:**

- Visual display output  
- Early operator interaction  

---

## Interactive Graphics: Foundational Breakthrough

### Sketchpad System (1963)

Developed by **Ivan Sutherland**, this marks the origin of modern interactive graphics.

#### Core Contributions

- Hierarchical data structures for graphical objects  
- Object instancing (reuse of components)  
- Interactive input using:
  - Keyboard  
  - Light pen  

#### Conceptual Innovations

- Direct manipulation of graphical objects  
- Constraint-based design  
- Real-time interaction  

#### Significance

Established principles still used in:

- CAD systems  
- Modern graphics APIs  
- Interactive design environments  

---

## Emergence of CAD/CAM Systems (1960s)

Graphics began to be applied in **engineering design**.

### Key Systems

- General Motors DAC (automobile design)  
- Itek Digitek (optical/lens design)  

### Functional Impact

- Enabled iterative design processes  
- Introduced interactive modelling  

### Limitation

- Expensive  
- Limited to specialized organizations  

---

## Batch Processing Era Constraints

### Characteristics

- Input/output via:
  - Punched cards  
- Minimal interactivity  

### Expectation

Graphics was envisioned as a tool for:

- Real-time human–computer interaction  

### Reality

Limited adoption due to:

- High hardware cost  
- Low computational resources  

---

## Expansion with Personal Computing (1980s)

### Key Transition

Introduction of personal computers with raster displays:

- Apple Macintosh  
- IBM PC  

---

### Core Technology: Bitmap Graphics

- Screen represented as a **grid of pixels (picture elements)**  
- Each pixel encoded as **binary or multi-bit values**  

### Impact

Rapid growth of:

- Graphical user interfaces  
- Low-cost applications (word processors, spreadsheets)  

---

## Desktop Metaphor

A conceptual model for organizing graphical interfaces.

### Components

- Windows (independent workspaces)  
- Icons (object representations)  
- Mouse (pointing device)  

### Functional Capabilities

- Overlapping windows  
- Multi-tasking via visual selection  
- Direct manipulation (click, drag, select)  

### Conceptual Shift

- From **command-line interaction** → **visual interaction paradigm**

---

## Direct Manipulation Interfaces

### Characteristics

Replace typed commands with:

- Pointing  
- Clicking  
- Dragging  

### Capabilities

- Object selection via icons  
- Menu-driven commands  
- Minimal reliance on keyboard input  

### Significance

- Lowered barrier to computer use  
- Enabled widespread adoption of graphics-based systems

# 1.2.1 Output Technology

## Vector (Random-Scan) Displays

### Terminology

- Vector / Stroke / Calligraphic displays  
- Lines drawn directly between endpoints  

### System Components

- Display processor  
- Display buffer  
- CRT  

### Operation

- Electron beam moves directly between points  
- Draws lines in arbitrary order  

### Key Property

- Requires continuous refreshing (~30 Hz) to prevent flicker  

### Advantages

- Smooth, continuous lines  
- High geometric precision  

### Limitations

- Poor support for filled areas  
- Limited realism  

---

## Raster Displays

### Core Concept

- Image represented as a **matrix of pixels (raster grid)**  

### System Components

- Frame buffer (pixel storage)  
- Video controller (image generation)  

### Operation

- Image scanned line-by-line (top to bottom)  
- Pixel intensity determines brightness/color  

---

### Color Representation

- **Grayscale:** 1 bit per pixel (bilevel)  
- **Low color:** 8 bits (256 colors)  
- **High color:** 24 bits (~16 million colors)  

---

### Memory Implications

Storage increases with:

- Resolution  
- Color depth  

---

## Vector vs Raster (Conceptual Comparison)

### Vector Systems

- Continuous representation  
- Direct line drawing  
- Limited realism  

### Raster Systems

- Discrete representation (pixels)  

Supports:

- Filled regions  
- Shading  
- Realistic images  

---

## Dominance of Raster Graphics

Driven by:

- Low hardware cost  
- Memory availability  
- Flexibility in image representation  

---

## Scan Conversion

### Definition

Process of converting:

- Geometric primitives (lines, polygons)  
→ Pixel representation  

### Importance

- Required for raster displays  
- Core operation in graphics pipelines  

---

## Aliasing (Sampling Artifact)

### Problem

- Discrete pixel grid approximates continuous shapes  

**Results in:**

- Jagged edges (“jaggies”)  

### Cause

- Undersampling of continuous signals  

### Solution

- Antialiasing techniques  
- Smooth intensity transitions  
- Reduced visual artifacts  

---

# 1.2.3 Software Portability and Graphics Standards

## Early Software Limitations

- Device-dependent systems  
- Complex and non-portable  
- High development cost  

---

## Device-Independent Graphics

### Concept

- Abstract graphics operations from hardware specifics  

### Benefits

- Portability across devices  
- Simplified programming  
- Reusability of applications  

---

## Standardization Efforts

### Core Graphics System (1977–1979)

- Baseline specification for graphics systems  

---

### GKS (Graphical Kernel System)

- Standardized 2D graphics system  
- Supports grouping via segments  

---

### GKS-3D

- Extension to 3D graphics  

---

### PHIGS (Programmer’s Hierarchical Interactive Graphics System)

#### Features

- Hierarchical structures  
- 3D transformations:
  - Scaling  
  - Rotation  
  - Translation  
- Retained-mode graphics (database-driven)  

#### Capability

- Automatic screen updates upon data changes  

---

### PHIGS PLUS

Extended with:

- Advanced rendering  
- Pseudo-realistic visualization  

---

## Industry (De Facto) Standards

### Characteristics

- Developed by companies/consortia  
- Rapid evolution  
- Strong commercial adoption  

### Examples

- PostScript  
- OpenGL  
- X Window System  

---

## Key Insight

Graphics software evolved from:

- Device-dependent → device-independent systems  
- Low-level control → high-level abstraction  

---

## Impact of This Shift

- Portability  
- Scalability  
- Widespread adoption of graphics applications  

---

## Summary

The historical evolution of computer graphics is driven by three interrelated factors:

1. **Hardware advancements** (CRT → raster displays)  
2. **Interaction improvements** (batch → interactive systems)  
3. **Software standardization** (device-dependent → portable systems)  

```python
# =========================================================
# 1. SETUP
# =========================================================

import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')


# =========================================================
# 2. CREATE FIGURE GRID (3 PER ROW)
# =========================================================

fig = plt.figure(figsize=(15, 10))


# =========================================================
# (1) VECTOR GRAPHICS (LINE DRAWING)
# =========================================================

ax1 = fig.add_subplot(2, 3, 1)

# Define line endpoints
x = [0, 1]
y = [0, 1]

ax1.plot(x, y)
ax1.set_title("Vector Graphics (Line)")
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)


# =========================================================
# (2) RASTER GRAPHICS (PIXEL GRID)
# =========================================================

ax2 = fig.add_subplot(2, 3, 2)

# Create pixel grid
grid = np.zeros((20, 20))

# Draw diagonal (pixel-based)
for i in range(20):
    grid[i, i] = 1

ax2.imshow(grid)
ax2.set_title("Raster Graphics (Pixel Grid)")
ax2.axis('off')


# =========================================================
# (3) SCAN CONVERSION (LINE → PIXELS)
# =========================================================

ax3 = fig.add_subplot(2, 3, 3)

# Discrete approximation of a line
x_vals = np.linspace(0, 10, 20)
y_vals = x_vals  # y = x

ax3.scatter(x_vals, y_vals)
ax3.set_title("Scan Conversion (Discrete Points)")
ax3.set_xlim(0, 10)
ax3.set_ylim(0, 10)


# =========================================================
# (4) ALIASING (JAGGED EFFECT)
# =========================================================

ax4 = fig.add_subplot(2, 3, 4)

# Low resolution sampling
x_alias = np.linspace(0, 10, 10)
y_alias = np.sin(x_alias)

ax4.plot(x_alias, y_alias)
ax4.set_title("Aliasing (Low Sampling)")


# =========================================================
# (5) ANTIALIASING (SMOOTH CURVE)
# =========================================================

ax5 = fig.add_subplot(2, 3, 5)

# High resolution sampling
x_anti = np.linspace(0, 10, 200)
y_anti = np.sin(x_anti)

ax5.plot(x_anti, y_anti)
ax5.set_title("Antialiasing (High Sampling)")


# =========================================================
# (6) GEOMETRIC TRANSFORMATION (BASIC GRAPHICS PIPELINE)
# =========================================================

ax6 = fig.add_subplot(2, 3, 6)

# Original square
square = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1],
    [0, 0]
])

# Rotation matrix
theta = np.pi / 4
R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

rotated_square = square @ R.T

ax6.plot(square[:, 0], square[:, 1], label="Original")
ax6.plot(rotated_square[:, 0], rotated_square[:, 1], linestyle='--', label="Rotated")

ax6.set_title("Geometric Transformation")
ax6.legend()
ax6.set_aspect('equal')


# =========================================================
# FINAL LAYOUT
# =========================================================

plt.tight_layout()
plt.show()
```

# 1.3 The Advantages of Interactive Graphics

Interactive graphics is a fundamental tool for visualization and human-computer interaction, leveraging the human brain’s natural aptitude for processing visual patterns. Its benefits span scientific, engineering, and design domains.

---

## 1.3.1 Cognitive and Perceptual Advantages

- Humans excel at **2D and 3D pattern recognition**, making graphical displays a natural interface for data interpretation.  
- Visual representations allow **rapid assimilation of complex information**, improving:
  - Decision-making  
  - Design evaluation  

### Examples of High-Value Visual Data

- Engineering schematics  
- Architectural designs  
- Scientific data from simulations or experiments  

---

## 1.3.2 Scientific Visualization

Scientific visualization emerged in the **late 1980s** as supercomputers began producing high-volume, multidimensional datasets.

### Capabilities of Graphics

- Summarization of large datasets  
- Detection of trends and patterns  
- Communication of abstract phenomena  

### Application Domains

- **Real-world objects:**
  - Aircraft  
  - Molecular structures  
  - Mechanical systems  

- **Abstract data:**
  - Population statistics  
  - Energy consumption  
  - Economic indicators  

---

## 1.3.3 Dynamic and Interactive Visualization

### Motion Dynamics

Objects can be:

- Rotated or tumbled relative to a fixed observer  
- Observed while the camera or viewer moves  

#### Applications

- Flight simulators (mechanical motion platforms with synchronized visual displays)  
- Interactive amusement-park rides  
- Video games exploiting 3D movement and dexterity challenges  

---

### Update Dynamics

Refers to **real-time changes in object properties**, such as:

- Shape  
- Color  
- Orientation  

#### Examples

- Deformations of an aircraft wing during flight  
- Real-time state changes in process control systems (e.g., nuclear reactors)  

### Benefits of Smooth Updates

- Improved realism  
- Better interpretability  
- Enhanced user engagement and situational awareness  

---

## 1.3.4 Enhanced Human-Computer Interaction

Interactive graphics provides a **high-bandwidth communication channel** between users and computers.

### Advantages

- Improved understanding of complex data structures  
- Ability to explore multiple perspectives dynamically  
- Facilitates:
  - Simulation  
  - Predictive modeling for real and synthetic systems  

> **Key Insight:**  
> Interactivity transforms passive observation into **active exploration**, which is critical for tasks requiring:
> - Visual reasoning  
> - Design iteration  
> - Decision support  

---

## 1.3.5 Summary

Interactive graphics is superior to static visualization due to its combination of:

- **Cognitive alignment** → leverages innate human visual processing  
- **Dynamic feedback** → allows motion and property updates in real time  
- **High-bandwidth interaction** → enables active exploration and control  

---

### Outcome

- Enhanced comprehension  
- Improved design efficiency  
- Stronger predictive analysis  

Applies to both:

- Concrete domains  
- Abstract domains  

```python
# =========================================================
# 1. SETUP
# =========================================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

plt.style.use('seaborn-v0_8')


# =========================================================
# 2. DATA PREPARATION
# =========================================================

# Dataset for scientific visualization
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Time variable for dynamic updates
t = np.linspace(0, 2*np.pi, 200)


# =========================================================
# 3. CREATE FIGURE GRID (3 PER ROW)
# =========================================================

fig = plt.figure(figsize=(15, 10))


# =========================================================
# (1) COGNITIVE ADVANTAGE (PATTERN RECOGNITION)
# =========================================================

ax1 = fig.add_subplot(2, 3, 1)

ax1.scatter(X[:, 2], X[:, 3], c=y)
ax1.set_title("Pattern Recognition (Clustering)")
ax1.set_xlabel("Petal Length")
ax1.set_ylabel("Petal Width")


# =========================================================
# (2) SCIENTIFIC VISUALIZATION (MULTI-DIMENSIONAL DATA)
# =========================================================

ax2 = fig.add_subplot(2, 3, 2, projection='3d')

ax2.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)
ax2.set_title("Scientific Visualization (3D)")
ax2.set_xlabel("Sepal L")
ax2.set_ylabel("Sepal W")
ax2.set_zlabel("Petal L")


# =========================================================
# (3) MOTION DYNAMICS (ROTATION)
# =========================================================

ax3 = fig.add_subplot(2, 3, 3)

# Circle rotating over time (simulated rotation)
x = np.cos(t)
y_rot = np.sin(t)

ax3.plot(x, y_rot)
ax3.set_title("Motion Dynamics (Rotation)")
ax3.axis('equal')


# =========================================================
# (4) UPDATE DYNAMICS (REAL-TIME CHANGE SIMULATION)
# =========================================================

ax4 = fig.add_subplot(2, 3, 4)

# Dynamic deformation (wave-like transformation)
x_wave = np.linspace(0, 10, 200)
y_wave = np.sin(x_wave) * np.exp(-0.1 * x_wave)

ax4.plot(x_wave, y_wave)
ax4.set_title("Update Dynamics (State Change)")


# =========================================================
# (5) MULTIPLE PERSPECTIVES (VIEW TRANSFORMATION)
# =========================================================

ax5 = fig.add_subplot(2, 3, 5)

# Same data, different projection
ax5.scatter(X[:, 1], X[:, 2], c=y)
ax5.set_title("Multiple Views (Projection)")
ax5.set_xlabel("Sepal Width")
ax5.set_ylabel("Petal Length")


# =========================================================
# (6) HIGH-BANDWIDTH VISUALIZATION (DENSE DATA)
# =========================================================

ax6 = fig.add_subplot(2, 3, 6)

# Dense synthetic dataset
dense_x = np.random.rand(1000)
dense_y = np.random.rand(1000)
colors = np.sqrt(dense_x**2 + dense_y**2)

ax6.scatter(dense_x, dense_y, c=colors)
ax6.set_title("High-Bandwidth Visualization")


# =========================================================
# FINAL LAYOUT
# =========================================================

plt.tight_layout()
plt.show()
```

# 1.4 Conceptual Framework for Interactive Graphics

Interactive graphics systems can be described using a **high-level conceptual framework** (Fig. 1.11). This framework separates **hardware, software, and user interaction responsibilities**.


image.png

---

## 1.4.1 System Components

### 1.4.1.1 Hardware Layer

Provides **input and output capabilities**:

- **Input devices:**
  - Mouse  
  - Data tablet  
  - Touch panel  
  - 3D locators  

- **Output devices:**
  - Monitors  
  - CRTs  
  - Raster or vector displays  

---

### 1.4.1.2 Software Layer

Divided into three components:

---

#### Application Program

- Handles user input  
- Updates the application model  
- Issues graphics commands  

**Responsibilities:**

- Specify what objects to display  
- Define how to respond to interactions  

---

#### Application Model

Represents all data and objects relevant to the display and post-processing modules.

**Types:**

- Explicit (e.g., CAD, scientific simulations)  
- Implicit (e.g., bitmap in painting programs)  

**Stores:**

- Geometric primitives:
  - Points, lines, curves, polygons, polyhedra, surfaces  
- Attributes:
  - Color, line style, textures  
- Connectivity and positioning information  
- Non-geometric properties:
  - Thermal, electrical, economic data  

---

#### Graphics System

- Converts model data into a visual representation for the display device  

**Functions:**

- Output transformation:
  - Model objects → rendered images  
- Input transformation:
  - User actions → model updates  

**Provides:**

- Logical abstraction of hardware  
- Primitives, attributes, and input sampling  

---

## 1.4.2 Application Modeling

### Purpose

Capture all **data, objects, and interrelationships** relevant to graphics or post-processing.

---

### Examples of Post-Processing Modules

- Simulation of circuits, aircraft stress analysis  
- Population or weather modeling  
- Financial and pricing computations  

---

### Model Representation

- Data arrays  
- Linked lists  
- Relational databases  
- Procedural definitions (e.g., fractals)  

**Key Properties:**

- Device-independent (independent of display device)  
- Often stored in an application database  

> Painting programs (e.g., MacPaint) may use the **bitmap itself as the model**.

---

## 1.4.3 Display of the Model

Model rendering is **separate from model creation**.

---

### Steps for Converting Model to Image

1. **Traversal**
   - Select geometric portions of the model  
   - Based on query or selection criteria  

2. **Conversion**
   - Translate extracted data and attributes into graphics commands  

---

### Primitives Supported for Output

- **2D:**
  - Lines, rectangles, polygons  
  - Circles, ellipses  
  - Text  

- **3D:**
  - Polygons  
  - Polyhedra  
  - Text  

---

### Graphics System Abstraction

- Provides a library of subroutines callable from:
  - C, Pascal, LISP  

- Shields programmers from:
  - Low-level hardware details  

- Treats input devices as:
  - Logical locators returning positions or event triggers  

---

## 1.4.4 Interaction Handling

### Event-Driven Loop

Standard paradigm for handling user interactions:

- Generate initial display from application model  
- Enable selection of commands or objects  
- Wait for user action  
- Process selection:
  - Update internal state  
  - Update model (if necessary)  
  - Update screen  
- Repeat until termination  

---

### Modes of Reaction

- **Screen-only updates:**
  - Highlight object  
  - Display menu  

- **Model updates:**
  - Add/remove components  
  - Modify attributes  

- Updates may involve:
  - Full redraw  
  - Incremental updates  

---

### Key Principle

> **Screen = window into the model**

- Model updates drive visual changes  
- Graphics system:
  - Does not modify the model  
  - Only renders it and passes input to the application program  

---

### Limitations

- Sequential (“ping-pong”) interaction  
- Parallel input/output (e.g., graphics + voice) is still an area of ongoing research  

---

## 1.4.5 Summary

Interactive graphics systems are **layered architectures** separating model, application, and hardware.

### Core Roles

- **Application program:** Defines objects and handles interactions  
- **Application model:** Represents abstract or concrete data independent of display  
- **Graphics system:** Renders views and abstracts hardware complexities  

---

### Final Insight

Event-driven loops provide structured **user–computer interaction**, central to all modern interactive systems.

```python
# =========================================================
# 1. SETUP
# =========================================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

plt.style.use('seaborn-v0_8')


# =========================================================
# 2. APPLICATION MODEL (REAL DATA)
# =========================================================

iris = datasets.load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names


# =========================================================
# 3. SIMULATED INTERACTION STATES
# =========================================================

# Define "user selections" (simulating interaction)
selected_class = 1   # user selects class 1
selected_feature_x = 2  # Petal Length
selected_feature_y = 3  # Petal Width


# =========================================================
# 4. CREATE FIGURE GRID (3 PER ROW)
# =========================================================

fig = plt.figure(figsize=(15, 10))


# =========================================================
# (1) INITIAL MODEL VIEW (ALL DATA)
# =========================================================

ax1 = fig.add_subplot(2, 3, 1)

ax1.scatter(X[:, 2], X[:, 3], c=y)
ax1.set_title("Initial View (Full Model)")
ax1.set_xlabel(feature_names[2])
ax1.set_ylabel(feature_names[3])


# =========================================================
# (2) FEATURE SELECTION (VIEW TRANSFORMATION)
# =========================================================

ax2 = fig.add_subplot(2, 3, 2)

ax2.scatter(X[:, selected_feature_x], X[:, selected_feature_y], c=y)
ax2.set_title("Feature Selection (User Input)")
ax2.set_xlabel(feature_names[selected_feature_x])
ax2.set_ylabel(feature_names[selected_feature_y])


# =========================================================
# (3) OBJECT SELECTION (FILTERING)
# =========================================================

ax3 = fig.add_subplot(2, 3, 3)

mask = (y == selected_class)
ax3.scatter(X[mask, 2], X[mask, 3])
ax3.set_title(f"Object Selection (Class {selected_class})")
ax3.set_xlabel(feature_names[2])
ax3.set_ylabel(feature_names[3])


# =========================================================
# (4) HIGHLIGHT SELECTION (SCREEN-ONLY UPDATE)
# =========================================================

ax4 = fig.add_subplot(2, 3, 4)

ax4.scatter(X[:, 2], X[:, 3], c=y)
ax4.scatter(X[mask, 2], X[mask, 3], s=100, edgecolors='black')

ax4.set_title("Highlight Selected Objects")
ax4.set_xlabel(feature_names[2])
ax4.set_ylabel(feature_names[3])


# =========================================================
# (5) MODEL UPDATE (SIMULATED TRANSFORMATION)
# =========================================================

ax5 = fig.add_subplot(2, 3, 5)

# Simulate model update (e.g., scaling selected data)
X_updated = X.copy()
X_updated[mask, 2] *= 1.2  # modify petal length

ax5.scatter(X_updated[:, 2], X_updated[:, 3], c=y)
ax5.set_title("Model Update (Data Modified)")
ax5.set_xlabel(feature_names[2])
ax5.set_ylabel(feature_names[3])


# =========================================================
# (6) FINAL RENDER (UPDATED VIEW)
# =========================================================

ax6 = fig.add_subplot(2, 3, 6)

ax6.scatter(X_updated[:, selected_feature_x],
            X_updated[:, selected_feature_y],
            c=y)

ax6.set_title("Final Rendered View")
ax6.set_xlabel(feature_names[selected_feature_x])
ax6.set_ylabel(feature_names[selected_feature_y])


# =========================================================
# FINAL LAYOUT
# =========================================================

plt.tight_layout()
plt.show()
```
